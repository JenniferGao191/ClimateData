```python
import pandas as pd
input_file = "Site_RW300new.csv"
# read the csv file
df = pd.read_csv(input_file)
df.columns = df.iloc[1]
df = df[3:]

# find the column index of 'Field-CH4' and 'Field-N2O' and store them in a list
col_index = [df.columns.get_loc(c) for c in df.columns if c in ['Field-CH4', 'Field-N2O']]
# change string to float
df.iloc[:, col_index] = df.iloc[:, col_index].astype(float)
df['No.  '] = df['No.  '].astype(int)
# group by column "No."
df_grouped = df.groupby("No.  ")
# sum the values of each group regarding 'Field-CH4' and 'Field-N2O' and generate a new dataframe
df_sum = df_grouped[['Field-CH4', 'Field-N2O']].sum().reset_index()
```

```python
# asending order according to the value of the No.
df_sum['No.  '] = df_sum['No.  '].astype(int)
df_sum.sort_values(by='No.  ', ascending=True)
df_sum.to_csv('df_sum.csv', index=False)
```
